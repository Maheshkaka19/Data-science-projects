```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\91879\Downloads\highest_gr_movies_2022.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df_twg = df.sort_values(by='world_gross',ascending=False)
```

```python
df_twg.head(10)
```

```python
sns.barplot(data=df_twg.head(10),y='Release',x='world_gross').set_title('Top ten movies by world collections')
```

```python
fig = px.bar(df_twg.head(10), y='Release', x='world_gross')
fig.update_layout(legend_orientation='h',
                  legend=dict(orientation='h',
                              x=1,
                              xanchor='right',
                              y=1.02,
                              yanchor='bottom',
                              traceorder='normal'),
                  xaxis={'categoryorder' : 'total ascending'},
                  title=('Top 10 Movies with world collections'))
fig.show()
```

```python
df_opc = df.sort_values(by = 'Opening',ascending=False)
```

```python
df_opc.head()
```

```python
fig = px.pie(df_opc.head(10),values='Opening',names='Release')
fig.update_layout(
    title=('Top 10 movies with highest opening day collections'))
fig.show()
```

```python
df_mth = df.sort_values(by = 'max_th',ascending=False)
```

```python
df_mth.head(10)
```

```python
sns.barplot(data = df_mth.head(10),x='max_th',y='Release').set(title='top 10 movies with max theaters')
```

```python
sns.pairplot(data=df)
```

```python
fig = px.histogram(df,x='Distributor',y='world_gross',
                   width=600,height=600,title='Distributors share of total collections')
fig.show()
```

```python
fig = px.treemap(df, path=[px.Constant('Rank'), 'Release', 'max_th'], values='world_gross',
                  color='Opening', hover_data=['Distributor'],width=800,height=800,title='Overall Data EDA')
fig.show()
```
